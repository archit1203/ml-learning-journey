<b>Rows: </b>Observation<br>
<b>Columns: </b>Features<br>

```python
import numpy as np
import pandas as pd
```

```python
#How to create a data frame => Numpy
# 1. Numpy Array
data = np.array([[1,4],[2,5],[3,6]])

# 2. Pandas Dataframe
#df = pd.DataFrame(data)
df = pd.DataFrame(data,index=['row1','row2',"row3"],columns=['col1','col2'])
print(df)
```

```python
#How to create a data frame => List
data = [[1,4],[2,5],[3,6]]

df = pd.DataFrame(data,index=['row1','row2',"row3"],columns=['col1','col2'])

df
```

```python
#How to create a data frame => Dictionary
states = ['UP', 'Tamil Nade', 'Sikkim','Maharashtra']
cap = ['Lucknow','Chennai','Gangtok','Mumbai'] 

dict_State = {'State':states, 'Capital':cap}

df_State = pd.DataFrame(dict_State)

df_State
```
