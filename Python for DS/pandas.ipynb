# 2D Array -> Dataframe

<b>Columns: </b>Features<br>

```python
import numpy as np
import pandas as pd
```

```python
#How to create a data frame => Numpy
# 1. Numpy Array
data = np.array([[1,4],[2,5],[3,6]])

# 2. Pandas Dataframe
#df = pd.DataFrame(data)
df = pd.DataFrame(data,index=['row1','row2',"row3"],columns=['col1','col2'])
print(df)
```

```python
#How to create a data frame => List
data = [[1,4],[2,5],[3,6]]

df = pd.DataFrame(data,index=['row1','row2',"row3"],columns=['col1','col2'])

df
```

```python
#How to create a data frame => Dictionary
states = ['UP', 'Tamil Nade', 'Sikkim','Maharashtra']
cap = ['Lucknow','Chennai','Gangtok','Mumbai'] 

dict_State = {'State':states, 'Capital':cap}

df_State = pd.DataFrame(dict_State)

df_State
```

```python
#How to create a data frame => CSV File
df_california = pd.read_csv('california_housing_test.csv')

#First 5 rows
df_california.head()

#Last 5 rows
df_california.tail()

#First N rows
df_california.head(10)
```

```python
df_california.shape
#(Rows,Colms)

#display all rows
pd.set_option('display.max_rows',3000)
df_california
```

## Pandas

Attributes: A value associated with an object => `df.columns`<br>
Function: A group of related statements that performs a specific task <br>
Methods: Functions which is defined inside a class body. eg `df.head()` <br>

```python
#Attributes
df_california.shape  #Shape -> (3000, 9)
df_california.index  #Index  -> RangeIndex(start=0, stop=3000, step=1)
df_california.columns #Columns ->Index(['longitude', 'latitude', 'housing_median_age', 'total_rooms','total_bedrooms', 'population', 'households', 'median_income','median_house_value'],dtype='object')
df_california.dtypes  #Data types of each column
```

```python
#Methods
df_california.head()  #1st 5 rows
df_california.info()  #Showing info of dataframe
df_california.describe() #Describe basic statistics of Dataframe
```

```python
#Functions

len(df_california) #Length
max(df_california) #Highest index
min(df_california) #Lowest index
type(df_california) #Data type
round(df_california,2) #Round off
```
