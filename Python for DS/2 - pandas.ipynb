## Adding a new Column

```python
#New column with a Scalar value
df_california['Country']='USA'
```

```python
#Adding a new column with Arrays => Numpy
#import numpy as np
score = np.arange(0,3000)
len(score)
df_california['Score'] = score
df_california.head()


#Random Integer => (inclusive,exclusice,size)
score2 = np.random.randint(0,100,3000)
len(score2)
df_california['Score']=score2
df_california.head()


#Random float 
np.random.uniform(1,100,3000)
```

```python
#Assign Method
#Add multiple columns with single line of code
#Override values of exisiting columns [BEST]
#It returns a new object/copy with all original columns in addition to new ones

score3=np.random.randint(0,100,3000)
score4=np.random.randint(0,100,3000)

series3=pd.Series(score3,index=np.arange(0,3000))
series4=pd.Series(score4,index=np.arange(0,3000))
```

```python
  #Using assign() to add multiple columns
  df_california.assign(score3=series3,score4=series4).head()
  df_california.head()  #Need to overrride
  df_california = df_california.assign(score3=series3,score4=series4)
  df_california.head()
```

## Insert 
insert a new column at a specific position

```python
df_california.insert(1,'test',series3)
df_california.head()
```

## Math Operations

```python
#Select a columm and find total sum
print('Sum: ',df_california['score3'].sum())
print('Count: ',df_california['score3'].count())
print('Mean: ',df_california['score3'].mean())
print('Std: ',df_california['score3'].std())
print('Max: ',df_california['score3'].max())
print('Min: ',df_california['score3'].min())
```

```python
#easier calculation with .describe()
df_california['score3'].describe()
df_california.describe()
```

```python
#operations in rows

#Sum in a row
(df_california['Score'] + df_california['score3'] + df_california['score4']).head()

#Calculating avg score & assign to new column
df_california['Avg'] = (df_california['Score'] + df_california['score3'] + df_california['score4'])/3

df_california.round(2)
df_california.head()
```

```python
#operations in rows

#Sum in a row
(df_california['Score'] + df_california['score3'] + df_california['score4']).head()

#Calculating avg score & assign to new column
df_california['Avg'] = (df_california['Score'] + df_california['score3'] + df_california['score4'])/3

df_california.round(2)
df_california.head()
```

```python
## Sort a Data Frame
```

```python
# Sort by 1 Column
df_california.sort_values(by='Score')
```

```python
#sort descending by one column
df_california.sort_values('Score',ascending=False)
```

```python
#sort descending by multiple column
#It doesn't make permanent changes it only views a copy
df_california.sort_values(['Score','Avg'],ascending=False)
```

```python
#sort descending by multiple column & update the data frame
#Inplace=True means update the data frame alse
df_california.sort_values(['Score','Avg'],ascending=False,inplace=True)
```

```python
#Sort descending with a key function
df_california.sort_values('Country',ascending=True,key=lambda col:col.str.lower())
```

## Create Index

```python
#import numpy as mp
import random

#Creating non repeating values for index
new_index = np.arange(0,3000)
```

```python
#Shuffling the indexes
random.shuffle(new_index)

#Create a new column with new index
df_california['New_Index']=new_index
```

```python
#Set new_Index column as index
df_california.set_index('New_Index',inplace=True)
```

```python
#Sort Index
df_california.sort_index(inplace=True)
```
