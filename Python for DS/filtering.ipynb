```python
df_laptops=pd.read_csv('laptops.csv')
df_laptops.head(3)
```

```python
#Find rows which have 'Apple' in the 'Brand' column
df_laptops['Brand']=='Asus'

#Filter Dataframe based on condition
df_laptops[df_laptops['Brand']=='Asus'].value_counts('Brand')
```
