# Create a Column Based on 1 Condition: np.where()

```python
#create an array based on price tiers
#np.where(df_laptops['Final Price']>2000,'Expensive','Cheap')

#Set it into a new Column
df_laptops['Price Tier'] = np.where(df_laptops['Final Price']>2000,'Expensive','Cheap')
df_laptops.value_counts('Price Tier')
df_laptops.head()
```

```python
#EXERCISE 1
#create an array based on Screen
#np.where(df_laptops['Screen']>14.0,'Big','Small')

df_laptops['Screen Size']=np.where(df_laptops['Screen']>14.0,'Big','Small')
df_laptops.value_counts('Screen Size')
```

# Creating a conditional column from more than 2 choices: np.select()

```python
#create an array based on multiple price tiers
#create conditions and values variable
conditions=[
    df_laptops['Final Price']>3000,
    (df_laptops['Final Price']>2000) & (df_laptops['Final Price']<=3000),
    (df_laptops['Final Price']>800) & (df_laptops['Final Price']<=2000),
    (df_laptops['Final Price']<=800)            
]

values=['Elite','Expensive','Budget','Cheap']

#set it to a new column
#np.select(conditions,values,default='Unknown')
df_laptops['Price Tier']=np.select(conditions,values,default='Unknown')

#show dataframe
df_laptops.head()

#count values inside price tier column
df_laptops.value_counts('Price Tier')
df_laptops['Price Tier'].value_counts()

df_laptops.head()
df_laptops['Screen'].value_counts()
```

```python
#EXERCISE 
#create an array based on mutliple screen tiers -> too big, big, small, too small

#create conditions and values variable
conditions=[ 
    (df_laptops['Screen']>16.0),
    (df_laptops['Screen']>14.0) & (df_laptops['Screen']<=16.0),
    (df_laptops['Screen']>12.0) & (df_laptops['Screen']<=14.0),
    (df_laptops['Screen']<=12.0)
]

values=['Large','Big','Regular','Small']

df_laptops['Screen Size']=np.select(conditions,values,default='UA')
#show dataframe
df_laptops.head()

#count values inside screen size column
df_laptops.value_counts('Screen Size') 
```

```python
# isin() : Single Filtering

#select Apple or HP laptops
df_laptops['Brand'].isin(['Apple','HP'])


df_laptops[df_laptops['Brand'].isin(['Apple','HP'])].value_counts('Brand')
df_laptops[df_laptops['Brand'].isin(['Apple','HP'])]
```
