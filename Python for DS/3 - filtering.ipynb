```python
import pandas as pd
import numpy as np
```

```python
df_laptops=pd.read_csv('laptops.csv')
df_laptops.head(3)
```

```python
#Find rows which have 'Apple' in the 'Brand' column
df_laptops['Brand']=='Asus'

#Filter Dataframe based on condition
df_laptops[df_laptops['Brand']=='Asus'].value_counts('Brand')
```

```python
#EXERCISE 1:
#Find rows which dont have 'HP' in the 'Brand' column
df_laptops['Brand']!='HP'

#Filter dataframe based on condition
df_laptops[df_laptops['Brand']!='HP'].value_counts('Brand')
```

```python
#EXERCISE 2:
#Find laptops with Final Price over 2000

df_laptops[df_laptops['Final Price'] > 2000].value_counts('Brand')
```

# Create a Column Based on 1 Condition: np.where()

```python
#create an array based on price tiers
#np.where(df_laptops['Final Price']>2000,'Expensive','Cheap')

#Set it into a new Column
df_laptops['Price Tier'] = np.where(df_laptops['Final Price']>2000,'Expensive','Cheap')
df_laptops.value_counts('Price Tier')
df_laptops.head()
```

```python
#EXERCISE 1
#create an array based on Screen
#np.where(df_laptops['Screen']>14.0,'Big','Small')

df_laptops['Screen Size']=np.where(df_laptops['Screen']>14.0,'Big','Small')
df_laptops.value_counts('Screen Size')
```

# Filer dataframe based on multiple conditions

```python
#Find Apple laptops that costs more than 2000
(df_laptops['Brand']=='Apple') & (df_laptops['Final Price']>2000)


df_laptops[(df_laptops['Brand']=='Apple') & (df_laptops['Final Price']>2000)]
```

```python
#Apple or Dell laptops
df_laptops[(df_laptops['Brand']=='Apple') | (df_laptops['Brand']=='Dell')]
df_laptops[(df_laptops['Brand']=='Apple') | (df_laptops['Brand']=='Dell')].value_counts('Brand')
```

```python
#Laptop from Apple or Dell that costs over 2000
df_laptops[((df_laptops['Brand']=='Apple') | (df_laptops['Brand']=='Dell')) & (df_laptops['Final Price']>2000)]
df_laptops[((df_laptops['Brand']=='Apple') | (df_laptops['Brand']=='Dell')) & (df_laptops['Final Price']>2000)].value_counts('Brand')
```
