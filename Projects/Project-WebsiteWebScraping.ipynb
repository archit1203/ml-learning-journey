```python
#Multiple Seasons
root="https://www.football-data.co.uk/mmz4281/"
leagues=['E0','E1','EC']

frames=[]
for league in leagues:
  for season in range(21,26):
    df = pd.read_csv(root+ str(season)+str(season+1) + '/' + league + '.csv',encoding='unicode_escape')
    df.insert(1,'Season',season)
    frames.append(df)
```

```python
# Organize All the data in dictionary
dict_league={
    'Spanish La Liga':'SP1',
    'Spanish Seguda Division': 'SP2',
    'German Bundesliga': 'D1',
    'English Premier League': 'E0',
    'English League 1':'E1'
}
```

```python
dict_historical_data={}


#Multiple Seasons
root="https://www.football-data.co.uk/mmz4281/"
#leagues=['E0','E1','EC']

for league in dict_league:
  frames=[]
  for season in range(20,26):
    df = pd.read_csv(root+ str(season)+str(season+1) + '/' + dict_league[league] + '.csv',encoding='unicode_escape')
    df.insert(1,'Season',season)
    frames.append(df)
  df_concat = pd.concat(frames)
  dict_historical_data[league]=df_concat
```
